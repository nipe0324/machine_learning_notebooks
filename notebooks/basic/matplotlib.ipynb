## Matplotlib
グラフ描画をするためのライブラリ。ディープラーニングでグラフの描画やデータの可視化をするときに使う。

```python
# jupyterで描画するためのおまじない
%matplotlib inline

# matplotlib をインポート
import numpy as np
import matplotlib.pyplot as plt

# データの作成
x = np.arange(0, 6, 0.1) # 0〜6までを0.1刻みで生成
y = np.sin(x) # 各xの値のsinを計算する

# グラフの描画
plt.plot(x, y)
plt.show()
```

```python
# 2つのグラフを描画
# データの作成
x = np.arange(0, 6, 0.1) # 0〜6までを0.1刻みで生成
y1 = np.sin(x)
y2 = np.cos(x)

# グラフの描画
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos', linestyle='--',)
plt.xlabel('x') # x軸のラベル
plt.ylabel('y') # y軸のラベル
plt.title('Plot sin & cos') # グラフのタイトル
plt.legend() # 凡例表示
plt.show()
```

```python
# 画像の表示
# 画像の読み込みは imread()、画像の表示は imshow() を利用します
import matplotlib.pyplot as plot
from matplotlib.image import imread

img = imread('./../images/earth.png') # 適切なパスを指定して読み込む
plt.imshow(img)
plt.show()
```
