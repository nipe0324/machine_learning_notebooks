```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_file = './../images/cat.jpeg'
jpeg_r = tf.read_file(img_file)
image = tf.image.decode_jpeg(jpeg_r, channels=3) # channel = 3 (rgb)
```

```python
sess = tf.Session()
init = tf.initialize_all_variables()
sess.run(init)
x = sess.run(image)
print(x.shape)
print(x)
```
