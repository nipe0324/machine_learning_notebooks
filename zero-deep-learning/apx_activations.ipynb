# 活性化関数(アクティベーション)

* ステップ関数: パーセプトロンで使われた
* シグモイド関数: ロジスティック回帰とか今までよく使われていた
* ReLU(Rectified Liner Unit): 最近DeepLearningで使われている

## ステップ関数

$$
\begin{array}{ll}
g(z) = \left\{ \begin{array}{ll}
0 & (z ≦ 0) \\
1 & (z > 0)  \\
\end{array} \right. \\
\\
z: 各入力と重みの総和 (W^TX) \\
\end{array}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# x軸の定義
z = np.arange(-5, 5, 0.1)

# ステップ関数
def step_function(z):
    y = z > 0;
    return y.astype(np.int)

y_step = step_function(z)

# グラフの描画
plt.plot(z, y_step)
plt.ylim(-0.2, 1.2)
plt.xlabel('z')
plt.ylabel('step(z)')
plt.title('Step Function')
plt.grid()
plt.show()
```

## シグモイド関数

$$
\begin{array}{ll}
g(z) = \frac{1} {1+ e^{-z} } \\
\\
z: 各入力と重みの総和 (W^TX) \\
\end{array}
$$


```python
# シグモイド関数
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

y_sig = sigmoid(z)

# グラフの描画
plt.plot(z, y_sig)
plt.ylim(-0.2, 1.2)
plt.xlabel('z')
plt.ylabel('sigmoid(z)')
plt.title('Sigmoid Function')
plt.grid()
plt.show()
```

## tanh

```python
y_tanh = np.tanh(z)

# グラフの描画
plt.plot(z, y_tanh)
plt.ylim(-1.5, 1.5)
plt.xlabel('z')
plt.ylabel('tanh(z)')
plt.title('tanh')
plt.grid()
plt.show()
```

## ReLU (Rectified Liner Unit)

$$
\begin{array}{ll}
g(z) = \left\{ \begin{array}{ll}
x & (z > 0) \\
0 & (z ≦ 0)  \\
\end{array} \right. \\
\\
z: 各入力と重みの総和 (W^TX) \\
\end{array}
$$


```python
# ReLU
def relu(z):
    return np.maximum(0, z)

y_relu = relu(z)

# グラフの描画
plt.plot(z, y_relu)
plt.ylim(-0.5, 5)
plt.xlabel('z')
plt.ylabel('ReLU(z)')
plt.title('ReLU')
plt.grid()
plt.show()
```
